```python
from linopy import Model
import warnings
warnings.filterwarnings("ignore")
```

```python
m = Model()
y = m.add_variables(binary = True, name = 'y')
x = m.add_variables(lower = 0, name = 'x')
```

```python
cons1 = m.add_constraints (-3*x+2*y>=1, name ="eq1")
cons2 = m.add_constraints (-8*x+10*y<=10, name ="eq2")
cons3 = m.add_constraints ( 0.3 <=x , name = 'eq3')
```

Maximize objective function

```python
obj = m.add_objective( x+y, sense="max")
```

```python
m.solve()
m.objective
```
